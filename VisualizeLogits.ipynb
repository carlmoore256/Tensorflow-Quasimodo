```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
file_dir = './saved_logits/'

files = glob.glob(file_dir + '*.npy')
print(len(files))
```

```python
for f in files:
    logit = np.load(f)
    print(logit[0].shape)
    plt.plot(logit[:])
    ipd.display(ipd.Audio(logit[:], rate=44100, autoplay=False))
    plt.show()
```
